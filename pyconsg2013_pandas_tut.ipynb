```python
def max_record(group, field=None):
    return group.ix[group[field].idxmax()]

def max_value(group):
    return max_record(group, 'value')

max_value(cafdata)
```

```python
cafdata.groupby('fgroup').apply(max_value)
```

```python
grouped = cafdata.groupby(['nutrient','fgroup'])
results = grouped.apply(max_value)
results
```
